```python
from matplotlib import pyplot as plt
```

```python
import pandas as pd
```

```python
data=pd.read_csv('data.csv')
```

```python
ids = data['Responder_id']
lang_responses = data['LanguagesWorkedWith']
```

```python
from collections import Counter
```

```python
language_counter=Counter()
```

```python
for response in lang_responses:
    language_counter.update(response.split(";"))
```

```python
languages=[item[0] for item in language_counter.most_common(5)]
```

```python
popularity=[item[1] for item in language_counter.most_common(5)]
```

```python
plt.pie(popularity,labels=languages)
plt.show()
```

```python
plt.pie(popularity,labels=languages,wedgeprops={'edgecolor':'black'})
plt.show()
```

```python
plt.pie(popularity,labels=languages,wedgeprops={'edgecolor':'black','linewidth':3},explode=[0.1,0,0,0,0],autopct='%1.1f%%',shadow=True)
plt.show()
```
