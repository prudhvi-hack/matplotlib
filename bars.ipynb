```python
from matplotlib import pyplot as plt
```

```python
from collections import Counter
```

```python
import csv
```

```python
with open('data.csv') as csv_file:
    csv_reader=csv.DictReader(csv_file)
    language_counter=Counter()
    for row in csv_reader:
        language_counter.update(row['LanguagesWorkedWith'].split(";"))
    
        
```

```python
language_counter.most_common(15)
```

```python
languages=[item[0] for item in language_counter.most_common(15)]

popularity=[item[1] for item in language_counter.most_common(15)]
```

```python
languages
```

```python
popularity
```

```python
plt.bar(languages,popularity)
plt.xlabel('languages')
plt.ylabel('popularity')
```

```python
plt.barh(languages,popularity)
```

```python
languages=languages[::-1]
popularity=popularity[::-1]
plt.barh(languages,popularity)
```

```python
import pandas as pd
```

```python
data=pd.read_csv('data.csv')
```

```python
ids = data['Responder_id']
lang_responses = data['LanguagesWorkedWith']
```

```python
language_counter=Counter()
```

```python
for responses in lang_responses:
    language_counter.update(responses.split(";"))
```

```python
language_counter=[item for item in language_counter.most_common(20) ]
```

```python
languages=[item[0] for item in language_counter]
```

```python
popularity=[item[1] for item in language_counter]
```

```python
plt.barh(languages[::-1],popularity[::-1])
plt.xlabel('popularity')
plt.show()
```
